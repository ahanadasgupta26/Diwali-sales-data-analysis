# EDA

#### GENDER

```python
df.columns
```

```python
ax = sns.countplot(x='Gender', data=df, hue='Gender', palette='Set2', legend=False)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount', data=sales_gen,hue='Gender', palette='Set2', legend=False)
```

### AGE

```python
ax = sns.countplot(x='Gender', data=df, hue='Gender', palette='Set2', legend=True)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount', data=sales_age,palette='Set2',hue='Age Group', legend=False)
```

### STATE

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Orders', data=sales_state,palette='Set2',hue='State', legend=False)
```

```python
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Amount', data=sales_state,palette='Set2',hue='State', legend=False)
```

### Marital Status

```python
ax = sns.countplot(x='Marital_Status', data=df, hue='Marital_Status', palette='Set2', legend=False)
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_status=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(x='Marital_Status',y='Amount', data=sales_status,hue='Gender',palette='Set2', legend=True)
```

### Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(x='Occupation', data=df, hue='Occupation', palette='Set2', legend=False)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occupation=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Occupation',y='Amount', data=sales_occupation,hue='Occupation',palette='Set2', legend=False)
```

### Product category

```python
sns.set(rc={'figure.figsize':(30,10)})
ax = sns.countplot(x='Product_Category', data=df, hue='Product_Category', palette='Set2', legend=False)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_cat=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(x='Product_Category',y='Amount', data=sales_cat,hue='Product_Category',palette='Set2', legend=False)
```

```python
sales_pid=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(x='Product_ID',y='Orders', data=sales_pid,hue='Product_ID',palette='Set2', legend=False)
```
